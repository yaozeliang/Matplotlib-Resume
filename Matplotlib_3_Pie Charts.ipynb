```python
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")


slices = [180,80,20,30]
labels = ('sixty','forty','extra1','extra2')
colors = ['#008fd5','#fc4f30','#e5ae37','#6d904f']

#  wedgeprops={'edgecolor':'black'} add the edgecolor
#  add labels 
plt.pie(slices,labels=labels,colors = colors,wedgeprops={'edgecolor':'black'})


plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()



# Colors:

# Blue = #008fd5

# Red = #fc4f30

# Yellow = #e5ae37

# Green = #6d904f
```

## Real example

```python

slices = [59219, 52466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

# Add shadow : shadow=True
# Rotate the chart: startangle=90
# Add percentage : autopct='%1.1f%%' show 1 digit after .
plt.pie(slices,labels=labels,shadow=True, 
        explode=explode,
        startangle=90, 
        autopct='%1.0f%%',
        wedgeprops={'edgecolor':'black'})


plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```
