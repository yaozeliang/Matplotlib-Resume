```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data.head()
```

## Fill the space under and above the median line

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')


plt.fill_between(ages, py_salaries,overall_median,where=(py_salaries>overall_median),interpolate=True,alpha=0.25,)


plt.fill_between(ages, py_salaries,overall_median,where=(py_salaries<overall_median),color = 'red',interpolate=True,alpha=0.25,)

plt.tight_layout()

plt.show()
```

## Fill between Python Dev and All Dev

```python



plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')




plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')


plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')


plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.tight_layout()

plt.show()
```
